# head=none is the condition of empty linkedlist


```python
from types import new_class
class LinkedList:
  def __init__(self):
    self.head=None
    self.n=0

  def __len__(self):
    return self.n

  def insert_head(self,data):
    new_node=Node(data)
    new_node.next=self.head
    self.head=new_node
    self.n+=1

  def __str__(self):
    curr=self.head
    result=' '

    while(curr != None):
      result= result + str(curr.data) + '->'
      curr=curr.next

    return result[:-2]



  def append(self,data):
    new_node=Node(data)
    if self.head==None:
      self.head=new_node
      self.n=self.n+1
      return


    new_node=Node(data)
    curr=self.head
    while curr.next !=None:
      curr=curr.next

    curr.next=new_node

  def insert_after(self,after,data):
    new_node=Node(data)
    curr=self.head

    while curr != None:
      if curr.data==after:
        break

      curr=curr.next

    #case1:break -> item found-> curr = not none
    if curr!=None:
      new_node.next = curr.next # The new node should point to what curr.next was pointing to
      curr.next=new_node       # curr.next should now point to the new node
      self.n=self.n+1
    #case2:not breal -> loop pura chala -> curr=none
    else:
      return "Item not Found"
```
