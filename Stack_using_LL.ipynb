```python
class Node():
  def __init__(self, data):
    self.data = data
    self.next = None
```

```python
class MyStack():
  def __init__(self):
    self.top=None
    self.size=0

  def isempty(self):
    return self.top==None


  def push(self,data):
    new_node=Node(data)
    if self.top==None:
      self.top = new_node # Corrected: Set the new node as the top

    else:
      new_node.next=self.top
      self.top=new_node
    self.size+=1 # size should increment regardless of whether it's the first element or not
    return self.size


  def traverse(self):
    temp=self.top
    if self.top==None:
      return " Empty Stack"

    else:
      while temp != None:
        print(temp.data)
        temp=temp.next

  def peek(self):
    if self.isempty():
      return "Empty Stack"
    return self.top.data


  def pop(self):
    if self.isempty():
      return "Empty Stack"

    else:
      data=self.top.data
      self.top=self.top.next
      return data



```

#String reversal using Stack

```python
def reverse_string(text):
  s=MyStack()
  for i in text:
    s.push(i)

  res=""

  while(not s.isempty()): # Corrected: Call isempty as a method
    res= res  + s.pop()

  return res
```
