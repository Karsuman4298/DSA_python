```python
import ctypes
```

```python
class Mylist():
  def __init__(self, size):
    self.size=size
    self.n=0
    self.A=self.__make_array(self.size)


  def __make_array(self,capacity):
    return(capacity*ctypes.py_object)()

  def __len__(self):
    return self.n


  def append(self,item):
    if self.size==self.n:
      #array is full-> resize
      self.__resize(self.size*2)

    self.A[self.n]=item
    self.n+=1

  def __resize(self,newcapacity):
    # Ensure a minimum capacity, especially if newcapacity becomes 0
    if newcapacity == 0:
        newcapacity = 1

    B=self.__make_array(newcapacity)

    for i in range(self.n):
      B[i]=self.A[i]

    self.A=B # Move this line outside the loop
    self.size=newcapacity


  def __str__(self):
    result=""
    for i in range(self.n):
      result=result + str(self.A[i]) + ","

    return "[" + result + "]"

  def __getitem__(self,index):
    if 0<=index<self.n:
      return self.A[index]

    else:
      return "Index Error"


  def pop(self):
    self.n=self.n-1
    return self.A[self.n]


  def clear(self):
    self.n=0
    self.size=1

  def find(self,item):
    for i in range(self.n):
      if(self.A[i]==item):
        return i
    return "Value Error"

  def insert(self,pos,item):
    if not (0 <= pos <= self.n):
        raise IndexError("Index out of bounds for insertion")

    if self.n == self.size:
      self.__resize(self.size * 2)

    # Shift elements to the right to make space for the new item
    for i in range(self.n, pos, -1):
      self.A[i] = self.A[i-1]

    self.A[pos] = item
    self.n += 1
```

```python

l=Mylist(0)
print(l.size)
print(type(l))
```
