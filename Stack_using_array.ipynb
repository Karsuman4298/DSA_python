```python
class Stack:
  def __init__(self,size):
    self.size=size
    self.stack= [None] * self.size
    self.top=-1

  def push(self,x):
    if self.top == self.size-1:
      return "Overflow"

    else:
      self.top=self.top+1
      self.stack[self.top]=x
```

```python
s=Stack(3)
```

```python
s.stack
```

```python
s.push(2)
```
