```python
class Stack:
  def __init__(self,size):
    self.size=size
    self.stack= [None] * self.size
    self.top=-1

  def push(self,x):
    if self.top == self.size-1:
      return "Overflow"

    else:
      self.top=self.top+1
      self.stack[self.top]=x

  def pop(self):
    if self.top==-1:
      return " Empty Stack"

    else:
      data=self.stack[self.top]
      self.top=self.top -1
      print(data)

  def traverse(self):
    for i in range(self.top+1):
      print(self.stack[i],end=' ')
```

```python
s=Stack(3)
```
