```python
class Node:
  def __init__(self, data):
    self.data = data
    self.next = None

```

```python
from os import spawnle
class Queue:

  def __init__(self):
    self.head=None
    self.rear=None


  def enque(self,value):
    new_node=Node(value)

    if self.rear == None:
      self.front=new_node
      self.rear=self.front


    else:
      self.rear.next=new_node
      self.rear=new_node


  def dequeue(self):
    if self.front== None:
      return " Empty"

    else:
      self.front=self.front.next


  def traverse(self):
    temp=self.front
    while temp != None:
      print(temp.data,end=" ")
      temp=temp.next



  def is_empty(self):
    return self.front==None

  def size(self):
    temp=self.front
    c=0

    while temp != None:
      c+=1
      temp=temp.next
    return c




```
