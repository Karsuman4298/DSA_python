```python
class Dictionary:
  def __init__(self,size):
    self.size=size
    self.slot=[None] * self.size
    self.data=[None] * self.size


  def put(self,key,value):
    hash_val=self.hash_func(key)

    if self.slot[hash_val] is None:
      self.slot[hash_val]=key
      self.data[hash_val]=value
    else:
      if self.slot[hash_val]==key:
        self.data[hash_val]=value
      else:
        new_hash_val=self.rehash(hash_val)

        while self.slot[new_hash_val] != None:
          new_hash_val=self.rehash(new_hash_val)




  def rehash(self,old_hash):
    return (old_hash + 1)% self.size

  def hash_func(self,key):
    return abs(hash(key)) % self.size

```
